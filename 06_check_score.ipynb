```python
# 성적 데이터 딕셔너리 (번호: [국어, 수학, 과학, 영어, 사회, 총평균])
scores_data = {
    1:  [85, 82, 93, 80, 96, 87.20],
    2:  [78, 75, 89, 72, 94, 81.60],
    3:  [92, 88, 96, 85, 98, 91.80],
    4:  [80, 76, 90, 77, 95, 83.60],
    5:  [75, 70, 88, 74, 92, 79.80],
    6:  [88, 85, 94, 82, 97, 89.20],
    7:  [70, 65, 85, 68, 90, 75.60],
    8:  [95, 92, 98, 90, 99, 94.80],
    9:  [82, 79, 91, 79, 95, 85.20],
    10: [76, 72, 89, 75, 93, 81.00],
    11: [84, 81, 92, 80, 96, 86.60],
    12: [79, 77, 90, 76, 94, 83.20],
    13: [87, 83, 95, 83, 97, 89.00],
    14: [81, 78, 91, 78, 94, 84.40],
    15: [73, 68, 87, 70, 91, 77.80],
    16: [90, 86, 97, 88, 98, 91.80],
    17: [68, 62, 84, 65, 89, 73.60],
    18: [86, 84, 93, 81, 96, 88.00],
    19: [77, 74, 89, 75, 93, 81.60],
    20: [83, 80, 92, 79, 95, 85.80]
}

print("=== 중학교 1학년 중간고사 성적 확인 ===")
print("조회하고 싶은 학생의 번호를 입력하세요 (종료하려면 'q' 입력)")

while True:
    user_input = input("\n번호 입력 > ")
    
    # 종료 조건
    if user_input.lower() == 'q':
        print("프로그램을 종료합니다.")
        break
        
    # 입력값 검증 및 처리
    try:
        student_num = int(user_input)
        
        if student_num in scores_data:
            # 해당 번호의 성적 데이터 가져오기
            s = scores_data[student_num]
            
            print(f"\n[ {student_num}번 학생 성적표 ]")
            print("------------------------")
            print(f"국어: {s[0]}점")
            print(f"수학: {s[1]}점")
            print(f"과학: {s[2]}점")
            print(f"영어: {s[3]}점")
            print(f"사회: {s[4]}점")
            print("------------------------")
            print(f"총평균: {s[5]}점")
            
        else:
            print(f"오류: {student_num}번 학생은 존재하지 않습니다. (1~20 사이 입력)")
            
    except ValueError:
        print("오류: 올바른 숫자를 입력해주세요.")
```
